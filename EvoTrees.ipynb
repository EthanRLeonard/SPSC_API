```julia
import Pkg
Pkg.add(url="https://github.com/Evovest/EvoTrees.jl")
```

```julia
using EvoTrees
x_train,y_train = rand(400,4),rand(400)

config = EvoTreeRegressor(
    loss=:linear, 
    nrounds=1000, 
    max_depth=8, 
    nbins=32,
    eta=0.1,
    lambda=0.1, 
    gamma=0.1, 
    min_weight=1.0,
    rowsample=0.5, 
    colsample=0.8)

m = fit_evotree(config; x_train, y_train)
preds = m(x_train) .- y_train
```

```julia
using Plots
plot(y_train)
```

```julia
plot(preds)
```
